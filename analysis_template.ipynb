## 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime, timedelta
import json

# Import our custom modules
from data_science_utils import DataScienceAnalyzer
from ai_engineering import VolunteerAIEngine

# Set visualization style
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (12, 6)

print("✅ Libraries imported successfully")
```

## 2. Initialize Data Analyzer

```python
# Initialize the analyzer
analyzer = DataScienceAnalyzer()
engine = VolunteerAIEngine()

print("✅ Analyzer initialized")
```

## 3. Data Quality Check

```python
# Check data quality
quality = analyzer.data_quality_report()

print("📊 Data Quality Report")
print("=" * 50)
print(f"\nUsers:")
print(f"  Total: {quality['users']['total']}")
print(f"  Admins: {quality['users']['admins']}")
print(f"  Volunteers: {quality['users']['volunteers']}")
print(f"  Missing email: {quality['users']['missing_email']}")

print(f"\nTasks:")
print(f"  Total: {quality['tasks']['total']}")
print(f"  Open: {quality['tasks']['open']}")
print(f"  Assigned: {quality['tasks']['assigned']}")
print(f"  Completed: {quality['tasks']['completed']}")

print(f"\nAttendance:")
print(f"  Total records: {quality['attendance']['total_records']}")
print(f"  Unique volunteers: {quality['attendance']['unique_volunteers']}")
```

## 4. Export Data

```python
# Export all data
users_df = analyzer.export_users()
tasks_df = analyzer.export_tasks()
attendance_df = analyzer.export_attendance()

print(f"✅ Data exported")
print(f"  Users: {len(users_df)} records")
print(f"  Tasks: {len(tasks_df)} records")
print(f"  Attendance: {len(attendance_df)} records")

# Display sample data
print("\n📋 Sample Users:")
print(users_df[['name', 'email', 'role', 'points']].head())
```

## 5. Engagement Analysis

```python
# Analyze engagement
engagement = analyzer.volunteer_engagement_analysis()

print("👥 Volunteer Engagement Analysis")
print("=" * 50)
summary = engagement['summary']
print(f"Total volunteers: {summary['total_volunteers']}")
print(f"Avg points: {summary['avg_points']:.1f}")
print(f"Avg attendance: {summary['avg_attendance']:.1f}")
print(f"Avg engagement score: {summary['avg_engagement_score']:.1f}/100")

print(f"\n📊 Engagement Distribution:")
for level, count in engagement['by_level'].items():
    print(f"  {level}: {count} volunteers")

print(f"\n🏆 Top 5 Performers:")
for i, performer in enumerate(engagement['top_performers'], 1):
    print(f"  {i}. {performer['name']}: {performer['engagement_score']:.1f}/100 ({performer['points']} points)")
```

## 6. Engagement Visualization

```python
# Visualize engagement
engagement_df = engagement['dataframe']

fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# Points distribution
axes[0, 0].hist(engagement_df['points'], bins=15, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Points Distribution')
axes[0, 0].set_xlabel('Points')
axes[0, 0].set_ylabel('Frequency')

# Attendance distribution
axes[0, 1].hist(engagement_df['attendance_count'], bins=15, color='lightgreen', edgecolor='black')
axes[0, 1].set_title('Attendance Distribution')
axes[0, 1].set_xlabel('Attendance Count')
axes[0, 1].set_ylabel('Frequency')

# Engagement score distribution
axes[1, 0].hist(engagement_df['engagement_score'], bins=15, color='coral', edgecolor='black')
axes[1, 0].set_title('Engagement Score Distribution')
axes[1, 0].set_xlabel('Engagement Score')
axes[1, 0].set_ylabel('Frequency')

# Engagement level pie chart
engagement_counts = engagement_df['engagement_level'].value_counts()
axes[1, 1].pie(engagement_counts.values, labels=engagement_counts.index, autopct='%1.1f%%')
axes[1, 1].set_title('Engagement Level Distribution')

plt.tight_layout()
plt.show()
```

## 7. Churn Risk Analysis

```python
# Predict churn risk
churn = engine.predict_churn_risk()

print("🚨 Churn Risk Analysis")
print("=" * 50)
print(f"Total volunteers: {churn['total_volunteers']}")
print(f"At-risk count: {churn['at_risk_count']}")
print(f"At-risk percentage: {churn['at_risk_percentage']:.1f}%")

print(f"\n📊 Risk Distribution:")
for level, count in churn['risk_distribution'].items():
    print(f"  {level}: {count} volunteers")

if churn['critical_risk']:
    print(f"\n🚨 Critical Risk Volunteers:")
    for volunteer in churn['critical_risk']:
        print(f"  - {volunteer['name']}: {volunteer['days_since_last_attendance']} days inactive")
```

## 8. Skill Demand Analysis

```python
# Analyze skill demand
skills = analyzer.skill_demand_analysis()

print("🎯 Skill Demand Analysis")
print("=" * 50)
print(f"Total unique skills: {skills['total_unique_skills']}")

print(f"\n📋 Skills with Gaps:")
for skill, data in list(skills['skills'].items())[:10]:
    if data['gap'] > 0:
        print(f"  {skill}:")
        print(f"    Demand: {data['demand']}, Supply: {data['supply']}, Gap: {data['gap']}")
        print(f"    Fulfillment: {data['fulfillment_rate']:.1f}%")
```

## 9. Volunteer Clustering

```python
# Cluster volunteers
clusters = engine.cluster_volunteers(n_clusters=3)

print("🎯 Volunteer Clustering")
print("=" * 50)
print(f"Number of clusters: {clusters['n_clusters']}")
print(f"Silhouette score: {clusters['silhouette_score']} (higher is better, max 1.0)")
print(f"Davies-Bouldin score: {clusters['davies_bouldin_score']} (lower is better)")

print(f"\n📊 Cluster Profiles:")
for profile in clusters['cluster_profiles']:
    print(f"\n  Cluster {profile['cluster_id']}:")
    print(f"    Size: {profile['size']} volunteers")
    print(f"    Avg points: {profile['avg_points']:.1f}")
    print(f"    Avg attendance: {profile['avg_attendance']:.1f}")
    print(f"    Avg engagement: {profile['avg_engagement']:.2f}")
```

## 10. Attendance Time Series

```python
# Analyze attendance trends
daily_attendance = analyzer.attendance_time_series(days=30)

if not daily_attendance.empty:
    plt.figure(figsize=(14, 6))
    plt.plot(daily_attendance['date'], daily_attendance['attendance_count'], marker='o', linewidth=2)
    plt.title('Daily Attendance (Last 30 Days)')
    plt.xlabel('Date')
    plt.ylabel('Attendance Count')
    plt.xticks(rotation=45)
    plt.grid(True, alpha=0.3)
    plt.tight_layout()
    plt.show()
    
    print(f"📅 Attendance Statistics (Last 30 Days)")
    print(f"  Total records: {daily_attendance['attendance_count'].sum()}")
    print(f"  Average per day: {daily_attendance['attendance_count'].mean():.1f}")
    print(f"  Max per day: {daily_attendance['attendance_count'].max()}")
    print(f"  Min per day: {daily_attendance['attendance_count'].min()}")
else:
    print("No attendance data available")
```
